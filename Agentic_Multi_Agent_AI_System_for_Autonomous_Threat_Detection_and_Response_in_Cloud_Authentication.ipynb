CELL 1: Detector Agent

```python
import random

def detect_anomaly(data):
    score = 0

    if data["location"] not in ["UK", "India"]:
        score += 0.4

    if data["device"] == "Unknown":
        score += 0.3

    hour = int(data["login_time"].split(":")[0])
    if hour < 6 or hour > 22:
        score += 0.3

    score += random.uniform(0, 0.1)

    return round(min(score, 1.0), 2)
```

CELL 2: Risk + Decision Agents

```python

def assess_risk(score):
    if score > 0.8:
        return "HIGH"
    elif score > 0.5:
        return "MEDIUM"
    else:
        return "LOW"


def make_decision(risk):
    if risk == "HIGH":
        return "BLOCK"
    elif risk == "MEDIUM":
        return "MFA_REQUIRED"
    else:
        return "ALLOW"
```

CELL 3: Logger Agent

```python
def log_result(data, score, risk, decision):
    log = {
        "user_id": data["user_id"],
        "score": score,
        "risk": risk,
        "decision": decision
    }
    return log
```

CELL 4: Run Full System

```python
def run_system(data):
    score = detect_anomaly(data)
    risk = assess_risk(score)
    decision = make_decision(risk)
    log = log_result(data, score, risk, decision)
    return log


sample_data = {
    "user_id": "U1001",
    "login_time": "02:30",
    "location": "Russia",
    "device": "Unknown"
}

run_system(sample_data)
```

```python
result = run_system(sample_data)

print("=== AUTONOMOUS SECURITY SYSTEM OUTPUT ===")
print(f"User ID: {result['user_id']}")
print(f"Anomaly Score: {result['score']}")
print(f"Risk Level: {result['risk']}")
print(f"Decision: {result['decision']}")
```
